```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from scipy.stats import pareto
from sklearn.cluster import KMeans
```

```python
pickle_in = open("./data_state/data.pickle", "rb")
df = pickle.load(pickle_in)
```

```python
df.tail()
```

```python
df['returned'] = df['InvoiceNo'].str.contains("C", na=False)
```

```python
df
```

```python
df = df[df['returned'] == False]
```

```python
df
```

## RECENCY

```python
currentDate = pd.to_datetime('2011-12-10')
```

```python
df['recency'] = currentDate - df['InvoiceDate']
```

```python
df
```

```python
df2 = pd.DataFrame({'customerID': df['CustomerID'].astype(int).unique().tolist()})
```

```python
df2 = df2.merge(df.groupby(['CustomerID']).agg({'recency':'min'}), left_on='customerID', right_on='CustomerID')
```

```python
df2['recency'] = df2['recency'] / pd.Timedelta(1, unit='d')
```

```python
df2
```

## FREQUENCY

```python
basket = df[['CustomerID', 'InvoiceNo']]
```

```python
df2 = df2.merge(basket.groupby(['CustomerID']).nunique(), left_on='customerID', right_on='CustomerID')
```

```python
df2.rename(columns={'InvoiceNo': 'frequency'}, inplace=True)
```

```python
df2
```

```python
# basket.groupby(['CustomerID']).nunique()
```

## Monetary

```python
basketUnits = df[['CustomerID', 'UnitPrice']]
```

```python
basketUnits.groupby(['CustomerID']).agg('count')
```

```python
df2 = df2.merge(basketUnits.groupby(['CustomerID']).agg('count'), left_on='customerID', right_on='CustomerID')
```

```python
df2.rename(columns={'UnitPrice': 'monetary'}, inplace=True)
```

```python
df2['monetary'] = df2['monetary'].apply(lambda x: 0 if (x < 0) else x) 
```

```python
total_max = df2['monetary'].sum()
```

```python
total_max
```

```python
df2['% of total'] = df2['monetary'].apply(lambda x: x / total_max * 100 ) 
```

```python
df2
```

## Pareto Principle

```python
df2.sort_values(by=['% of total'], ascending=False, inplace=True)
```

```python
df2['cumulative % total'] = df2['% of total'].cumsum(axis = 0)
```

```python
df2['rank'] = np.arange(len(df2))
```

```python
df2[df2['cumulative % total'] <= 20]
```

```python
df2.plot(kind='line', x='rank', y='cumulative % total', legend=None, title='Pareto Plot')
plt.show()
```

## k-means

```python
df2['frequency_z'] = np.log(df2['frequency'])
```

```python
df2['monetary_z'] = np.log(df2['monetary'])
```

```python
df2['recency_z'] = np.log(df2['recency'])
```

```python
df2
```

```python
km = KMeans(n_clusters=5, init='random', random_state=0)
```

```python
y = km.fit_predict(df2[['frequency_z', 'monetary_z', 'recency_z']])
```

```python
df2['cluster'] = y
```

```python
df2
```

# visualising the distribution in a 3d plot

```python
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")

ax.grid(b = True, color ='grey', 
        linestyle ='-.', linewidth = 0.3, 
        alpha = 0.2)
my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(df2['frequency_z'], df2['monetary_z'], df2['recency_z'], alpha = 0.8)
```

# visualising  the distribution by cluster

```python
def plotClusters(xAxis, yAxis):
    plt.scatter(df2[df2['cluster']==0][xAxis], df2[df2['cluster']==0][yAxis], s=40, c='red', label ='cluster 1')
    plt.scatter(df2[df2['cluster']==1][xAxis], df2[df2['cluster']==1][yAxis], s=40, c='blue', label ='cluster 2')
    plt.scatter(df2[df2['cluster']==2][xAxis], df2[df2['cluster']==2][yAxis], s=40, c='green', label ='cluster 3')
    plt.scatter(df2[df2['cluster']==3][xAxis], df2[df2['cluster']==3][yAxis], s=40, c='pink', label ='cluster 4')
    plt.scatter(df2[df2['cluster']==4][xAxis], df2[df2['cluster']==4][yAxis], s=40, c='gold', label ='cluster 5')
    plt.xlabel(xAxis)
    plt.ylabel(yAxis)    
    plt.legend() 
    
plotClusters('frequency_z', 'monetary_z')
```
