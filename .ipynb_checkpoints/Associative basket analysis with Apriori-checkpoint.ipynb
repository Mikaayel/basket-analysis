```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('./data/Groceries_dataset.csv');
```

```python
df.head()
```

```python
df['itemDescription'] = df['itemDescription'].str.strip();
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df.itemDescription)
```

```python
dummies
```

```python
mergedDF = pd.concat([df, dummies], axis='columns')
```

```python
mergedDF.pop('itemDescription')
```

```python
mergedDF
```

```python
basket = mergedDF.groupby(['Member_number', 'Date'], as_index=False).sum()
```

```python
basket.pop('Member_number')
basket.pop('Date')
basket.pop('other vegetables')
```

```python
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.tail()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
```

```python
frequent_itemsets.head(20)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.0001)
```
